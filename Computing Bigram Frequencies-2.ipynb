# Exercise-1:Process simple bigram data file

# Step 1:OPEN the file,count_2w.txt

```python
file=open(r"count_2w.txt",encoding="UTF-8")
a=file.readlines()
a
```

# Step 2: build goog2w_list

```python
mini=a[:10]
mini[0]
```

```python
mini[0].split()
```

```python
mini_list=[]
for m in mini:
    (w1,w2,count)=m.split()
    count=int(count)
    mini_list.append(((w1,w2),count))
mini_list
```

```python
mini_list[0] 
```

# Step 3: Build goog2w_fd

```python
import nltk 
goog2w_fd=nltk.FreqDist() 
for m in a: 
    w1,w2,count=m.split()  
    goog2w_fd[(w1,w2)]=count 
```

```python
goog2w_fd[('of','the')] 
```

```python
goog2w_fd[('so','beautiful')] 
```

# Step 4: Explore

# 1.What are the top-10 bigrams?

```python
goog2w_fd.most_common(10) 
```

# 2.What are the top so-initial bigrams?

```python
text=[]
for t in a: 
    (w1,w2,count)=t.split() 
    count=int(count) 
    if w1=='so': 
        text.append(((w1,w2),count))   
text
```

# Step 5: Pickle the data

```python
import pickle 
with open('goog2w_list.pkl','wb') as handle: 
    pickle.dump(goog2w_fd,handle,protocol=pickle.HIGHEST_PROTOCOL)
```

# Exercise 3: Bigram frequencies of Jane Austen Novels 

```python
with open("austen-emma.txt") as fn: 
    nov=fn.read(500) 
    print(nov) 
```

```python
tokenizer=nltk.tokenize.WhitespaceTokenizer() 
tok=tokenizer.tokenize(nov) 
print(tok)
```

# 1.a_toks:word tokens, all in lowercase 

```python
tokenizer=nltk.tokenize.WhitespaceTokenizer() 
a_tok=tokenizer.tokenize(nov.lower()) 
print(a_tok)
```

```python
b2=list(nltk.bigrams(tok)) 
b2fd=nltk.FreqDist(b2) 
b2fd 
```

```python
import re 
from collections import Counter
```

```python
words=re.findall(r'so+ \w+',open('austen-emma.txt').read()) 
ab=Counter(zip(words)) 
print(ab)
```

# 2.a_tokfd:word frequency distribution

```python
a_tokfd=FreqDist(a_tok) 
a_tokfd
```

# 3.a_bigrams:word bigrams,cast as a list 

```python
a_bigrams=list(nltk.bigrams(a_tok)) 
a_bigrams
```

# 4.a_bigramfd:bigram frequency distribution

```python
a_bigramfd=nltk.FreqDist(a_bigrams) 
a_bigramfd 
```
